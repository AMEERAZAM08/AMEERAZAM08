```python
import dlib  
import cv2
import matplotlib.pyplot as plt
import mediapipe  as mp
import pandas as pd  
import  glob
import  numpy as np 
detector=dlib.get_frontal_face_detector()
predictor=dlib.shape_predictor("../input/shape-predictor-68-face-landmarks/shape_predictor_68_face_landmarks.dat") 
mp_face_detection = mp.solutions.face_detection
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(min_detection_confidence=0.5, min_tracking_confidence=0.5)
np.random.seed(0)
```

```python
def  pose_found(image):
    with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
        success, image = True,image#cv2.imread(image)
        find_pose =face_mesh.process(image)

        img_h, img_w, img_c = image.shape
        face_3d = []
        face_2d = []

        if find_pose.multi_face_landmarks:

            for face_landmarks in find_pose.multi_face_landmarks:
                # print('*')
                for idx, lm in enumerate(face_landmarks.landmark):
                    if idx == 33 or idx == 263 or idx == 1 or idx == 61 or idx == 291 or idx == 199:
                        x, y = int(lm.x * img_w), int(lm.y * img_h)
                        face_2d.append([x, y])
                        face_3d.append([x, y, lm.z])       
                
                face_2d = np.array(face_2d, dtype=np.float64)
                face_3d = np.array(face_3d, dtype=np.float64)

                # The camera matrix
                focal_length = 1 * img_w

                cam_matrix = np.array([ [focal_length, 0, img_h / 2],
                                        [0, focal_length, img_w / 2],
                                        [0, 0, 1]])

                dist_matrix = np.zeros((4, 1), dtype=np.float64)
                success, rot_vec, trans_vec = cv2.solvePnP(face_3d, face_2d, cam_matrix, dist_matrix)
                rmat, jac = cv2.Rodrigues(rot_vec)
                angles, mtxR, mtxQ, Qx, Qy, Qz = cv2.RQDecomp3x3(rmat)
                # Get the y rotation degree
                x = angles[0] * 360
                y = angles[1] * 360
                z = angles[2] * 360  #is for   z rotation degree 
                
                angle_x=round(x,3)
                angle_y=round(y,3)
                angle_z=round(z,3)
    # plt.imshow(image)
    # print(angle_x,angle_y,angle_z)
    return angle_x,angle_y,angle_z

# pose_found(angry[0])
```

```python
from scipy.spatial import distance as dist
import  tqdm
def eye_aspect_ratio(eye):
    p2_minus_p6 = dist.euclidean(eye[1], eye[5])
    p3_minus_p5 = dist.euclidean(eye[2], eye[4])
    p1_minus_p4 = dist.euclidean(eye[0], eye[3])
    ear = (p2_minus_p6 + p3_minus_p5) / (2.0 * p1_minus_p4)
    return ear
```
